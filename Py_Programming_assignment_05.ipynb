1. Write a Python Program to Find LCM?

```python
def lcm(a,b):
      for i in range(max(a, b), 1 + (a * b)):
          if i % a == i % b == 0:
              c = i
              break
      print(f"LCM of {a} and {b} is {c}")


lcm(4,8)
```

5. Write a Python Program to Make a Simple Calculator with 4 basic mathematical operations?

```python
def add(a,b):
  c = a + b
  print(f"Addition of two number is: {c}")


def subtract(a,b):
  c = a - b
  print(f"Subtraction of two number is: {c}")


def multiply(a,b):
  c = a * b
  print(f"Multiplication of two number is: {c}")


def devision(a,b):
  c = a / b
  print(f"Devision of two number is: {c}")


add(40,8)
subtract(40,8)
multiply(40,8)
devision(40,8)
```
