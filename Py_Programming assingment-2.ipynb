1. Write a Python program to convert kilometers to miles?

```python
# 1 km = 0.62137119 miles

def kilo_mile(A):
   miles = A * 0.62137119
   print(f"{A} kilometer is equal to {miles} in miles.")


   
kilo_mile(2)
```

2. Write a Python program to convert Celsius to Fahrenheit?

```python
#(°C × 9/5) + 32 = °F

def cel_fahren(A):
  B = (A * (9/5)) + 32
  print(f"{A} Celsius in {B} Fahrenheit")


cel_fahren(38)
```
